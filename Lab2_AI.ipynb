```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/heart.csv")
df.head(5)
```

```python
# Categorical data
# countplot
sns.countplot(data=df, x='sex')
df['fbs'].value_counts()
```

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

```python
plt.hist(df['age'])
```

```python
# Distplot
# probability density function --> The line is kde
# skweness
sns.distplot(df['age'])
```

```python
# Boxplot
# 5 number summary
# median , percentile, min=Q1-1.5*IQR, max=Q3+1.5*IQR, IQR = (Q1-Q3), outlier

sns.boxplot(df['age'])
```

```python
df.describe()
```

```python
sns.scatterplot(x='age', y='cp',hue='sex', style='trtbps',size='chol' ,data=df.head())
plt.show()
```

```python
 # Bar plot (Numerical - Categorical)
  # average age in pclass
sns.barplot(x='cp',y='age',hue='sex',data=df)
```

```python
#Box plot (Numerical - Categorical)

sns.boxplot(x='sex',y='age',hue='cp',data=df)
```

```python
sns.distplot(df[df['fbs']==0]['age'],hist=False)
sns.distplot(df[df['fbs']==1]['age'],hist=False)
```

```python
# HeatMap (Categorical - Categorical)
sns.heatmap(pd.crosstab(df['cp'],df['fbs']))
```

```python
df.groupby('cp').mean()['fbs']*100
```

```python
# pairplot
sns.pairplot(df.head(),hue='restecg')
```

```python
sns.lineplot(x='age',y='sex',data=df.head())
```

```python
sns.heatmap(df.pivot_table(values='age',index='cp',columns='sex'))
```
